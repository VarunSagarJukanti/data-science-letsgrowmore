```python
import matplotlib.pyplot as plt
from skimage import io
import cv2
from google.colab.patches import cv2_imshow
```

```python
from skimage.util.dtype import img_as_float
img_as_float = io.imread("https://www.grange.co.uk/upload/rolls-royce-cullinan-exterior-design.jpg")
```

```python
cv2_imshow(img_as_float)
```

```python
#gray filter
gray= cv2.cvtColor(img_as_float, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray)
```

```python
#inverted filter
inverted= cv2.bitwise_not(gray)
cv2_imshow(inverted)
```

```python
#image smoothing
image_smoothing = cv2.GaussianBlur(inverted, (21,21), sigmaX=0, sigmaY=0)
cv2_imshow(image_smoothing)
```

```python
#penclsketch
def pencil_sketch(x, y):
  return cv2.divide(x, 255 - y, scale=256)
pencil_sketch = pencil_sketch(gray, image_smoothing)
cv2_imshow(pencil_sketch)
```
