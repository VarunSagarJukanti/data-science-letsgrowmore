```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv(r'/content/globalterrorismdb_0718dist.csv',encoding="latin")
```

```python
data.describe()
```

```python
data.head()
```

```python
from pandas.io.formats.format import ColspaceArgType
col= data.columns.to_list()
col
```

```python
copy_data = pd.DataFrame(data, data.rename(columns={ 'success': 'Success', 'location': 'Location', 'iyear':'Year','imonth':'Month','extended':'Extended','iday':'Day','country_txt':'Country',
                   'country': 'Country_code', 'provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
                   'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                   'weaptype1_txt':'Weapon_type','motive':'Motive','city':'City','latitude':'Latitude','longitude':'Longitude'},inplace=True))
```

```python
df=copy_data['City'].value_counts().to_frame().drop('Unknown').sort_values('City',axis=0,ascending=False).head(10)
df.plot(kind='bar',figsize=(20,10),color='darkgreen')
plt.xticks(rotation = 50)
plt.xlabel("name of the City",fontsize=20,color="violet")
plt.ylabel("Total attacks in a city",fontsize=20,color="indigo")
plt.title("Top 10 most affected cities",fontsize=20,color="skyblue")
plt.show()
```

```python
plt.subplots(figsize=(20,10))
index1=copy_data['Country'].value_counts()[:15].index
values1=copy_data['Country'].value_counts()[:15].values
sns.barplot(index1,values1,palette='rocket')
plt.title('Top 10 most affected countries', fontsize = 50,color="blue")
plt.xlabel('name of the Country', fontsize = 25,color="red")
plt.ylabel('Total attacks in an country', fontsize = 35,color="magenta")
plt.xticks(rotation = 50)
plt.show()
```

```python
df1=copy_data['Region'].value_counts().to_frame().sort_values('Region',axis=0,ascending=False).head(10)
df1.plot(kind='bar',figsize=(20,10),color='blue')
plt.xticks(rotation = 50)
plt.xlabel("name of the Region",fontsize=20,color="magenta")
plt.ylabel("Total attacks",fontsize=20)
plt.title("Top 10 most affected regions",fontsize=20,color="gray")
plt.show()
```

```python
df2=copy_data[['Group','Killed']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('Killed',ascending=False).head(10)
df2.plot(kind='bar',color='grey',figsize=(20,10))
plt.title("Top 10 terrorist attack groups",fontsize=20,color="green")
plt.xlabel("Terrorist group",fontsize=30,color="yellow")
plt.ylabel("Total attacks",fontsize=40,color="pink")
plt.xticks(rotation=50)
plt.show()
```

```python
copy_data['AttackType'].value_counts().plot(kind='bar',figsize=(20,10),color='pink')
plt.xticks(rotation = 50)
plt.xlabel("Attack Type",fontsize=20)
plt.ylabel("Total attacks",fontsize=30)
plt.title("Most used attack types",fontsize=20)
plt.show()
```
